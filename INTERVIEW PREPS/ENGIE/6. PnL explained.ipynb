# PnL Explained : Définition et Utilisation des Greeks et Séries de Taylor dans le Contexte d'ENGIE

## 1. Qu'est-ce que le PnL Explained ?

Le **PnL Explained** est une décomposition détaillée des profits et pertes (PnL) en fonction des facteurs de marché et des paramètres clés. Il permet de comprendre précisément pourquoi une position ou un portefeuille a généré des gains ou des pertes. 

### Objectifs du PnL Explained :
- Identifier les contributions des variations de marché (prix, volatilité, taux).
- Expliquer les écarts entre le PnL réalisé et attendu.
- Évaluer les impacts des décisions de gestion ou des ajustements de portefeuille.

---

## 2. Méthodes d’Analyse : Greeks et Séries de Taylor

### a. Les Greeks
Les **Greeks** sont des sensibilités qui mesurent comment le PnL d’une position évolue en fonction de différents facteurs de marché. Ces indicateurs sont particulièrement utiles dans le cadre de la gestion des produits dérivés.

#### Principaux Greeks :
1. **Delta (Δ)** :
   - Sensibilité du PnL aux variations du prix de l’actif sous-jacent.
   - **Exemple ENGIE** : Une position long sur le gaz naturel aura un Delta positif : une hausse des prix du gaz augmentera le PnL.

2. **Gamma (Γ)** :
   - Sensibilité du Delta aux variations du prix de l’actif sous-jacent.
   - **Exemple ENGIE** : Utilisé pour évaluer les risques de non-linéarité dans les contrats d'options énergétiques.

3. **Vega (ν)** :
   - Sensibilité du PnL à la volatilité du marché.
   - **Exemple ENGIE** : Une option sur le prix de l’électricité devient plus précieuse lorsque la volatilité des prix de l’électricité augmente.

4. **Theta (Θ)** :
   - Sensibilité du PnL à l’écoulement du temps (décay time).
   - **Exemple ENGIE** : Une option proche de l’expiration perd de la valeur avec le temps si le marché reste stable.

5. **Rho (ρ)** :
   - Sensibilité du PnL aux variations des taux d'intérêt.
   - **Exemple ENGIE** : Impact des variations des taux d’actualisation sur les contrats de financement énergétique.

---

### b. Séries de Taylor
La **série de Taylor** permet d'approximer les variations du PnL en fonction des multiples sensibilités (Greeks) de manière plus détaillée. 

#### Formule Générale :
\[
PnL = Δ \cdot \Delta S + \frac{1}{2} Γ \cdot (\Delta S)^2 + ν \cdot \Delta \sigma + Θ \cdot \Delta t + ρ \cdot \Delta r + \text{termes résiduels}
\]

- \( \Delta S \) : Variation du prix du sous-jacent.
- \( \Delta \sigma \) : Variation de la volatilité.
- \( \Delta t \) : Temps écoulé.
- \( \Delta r \) : Variation des taux d’intérêt.

#### Application chez ENGIE :
ENGIE peut utiliser une approximation Taylor pour :
- Modéliser l’impact des variations des prix du gaz et de l’électricité sur son portefeuille de dérivés.
- Analyser les impacts combinés des mouvements de marché sur les options énergétiques.

---

## 3. Contexte d'ENGIE : Exemple de PnL Explained

### Cas : Gestion d'une Option sur le Gaz Naturel
#### Données :
- **Position** : Option call sur le gaz naturel.
- **Prix initial du gaz** : 50 €/MWh.
- **Prix actuel** : 55 €/MWh.
- **Volatilité** : 25 %.
- **Greeks calculés** :
  - \( \Delta = 0.6 \),
  - \( Γ = 0.1 \),
  - \( ν = 0.5 \),
  - \( Θ = -0.02 \).

#### Analyse PnL avec Séries de Taylor :
1. **Impact Delta** :
   \[
   PnL_{\Delta} = 0.6 \cdot (55 - 50) = 3 €/MWh
   \]

2. **Impact Gamma** :
   \[
   PnL_{\Gamma} = \frac{1}{2} \cdot 0.1 \cdot (55 - 50)^2 = 1.25 €/MWh
   \]

3. **Impact Vega (volatilité)** :
   \[
   PnL_{\nu} = 0.5 \cdot \Delta \sigma = 0.5 \cdot (30\% - 25\%) = 0.25 €/MWh
   \]

4. **Impact Theta (décay time)** :
   \[
   PnL_{\Theta} = -0.02 \cdot 1 \, \text{jour} = -0.02 €/MWh
   \]

#### Résultat Total :
\[
PnL = PnL_{\Delta} + PnL_{\Gamma} + PnL_{\nu} + PnL_{\Theta} = 3 + 1.25 + 0.25 - 0.02 = 4.48 €/MWh
\]

ENGIE peut ainsi expliquer une variation positive de 4.48 €/MWh dans son PnL en fonction des contributions de chaque facteur de marché.

## 4. Importance pour ENGIE

1. **Transparence et Contrôle** :
   - Le PnL Explained aide ENGIE à comprendre les sources de profits ou pertes dans ses activités de trading énergétique.
   
2. **Gestion des Risques** :
   - Les Greeks permettent d’identifier les expositions principales et de mettre en place des couvertures (hedging) adaptées.

3. **Optimisation des Portefeuilles** :
   - Les analyses basées sur les séries de Taylor permettent de modéliser différents scénarios de marché pour ajuster les stratégies.

---

## 5. Conclusion

Le PnL Explained, en s'appuyant sur les Greeks et les séries de Taylor, est un outil indispensable pour ENGIE dans la gestion de ses positions et portefeuilles énergétiques. Il permet une analyse fine des impacts de marché, favorisant une meilleure gestion des risques et une optimisation de la performance.
